**Задача 3:**

С помощью готовой функции (пример ниже) создайте по 2-м точкам кубический сплайн, задав в качестве граничных условий значения 1-й или 2-й производной многочлена на каждом из концов. Убедитесь, что сплайн совпадает с исходным многочленом, построив их графики.

````python

f = lambda x: 1 / (25*x**2 + 1)
n = 10
x = np.linspace(-1,1,10)
y = f(x)
cs = CubicSpline(x, y, bc_type = 'natural') # Нужно поменять г.у. здесь
xs = np.linspace(-1.1,1.1,100)
plt.figure(figsize=(6.5, 4))
plt.plot(x, y, 'o', label='data')
plt.plot(xs, f(xs), label='true')
plt.plot(xs, cs(xs), label="S")
plt.xlim(-1, 1)
plt.legend()
````
