**Задача 1**  
Пусть $\Vert x\Vert$ - векторная норма в $\mathbb{C}^n$, $A \in \mathbb{C}^n$ - невырожденная матрица. Докажите, что $\Vert x\Vert_{*} = \Vert A x\Vert$  также является векторной нормой.


**Решение**  
Проверим, что выполняются все свойства нормы (везде пользуемся тем, что для $||\cdot||$ все выполняется):
1. $\Vert x \Vert_* = 0 = \Vert A x \Vert = 0 \Rightarrow$ (т.к. $||\cdot||$ это норма) $Ax = 0 \Rightarrow$ (т.к. матрица $A$  невырожденная) $x = 0$. В обратную сторону тоже выполняется.

2. $\Vert a x \Vert_* = || a A x|| = |a| ||A x|| = |a| ||x||_*$

3. $||x + y||_* = ||A(x+y)|| \le ||Ax|| + ||Ay|| = ||x||_* + ||y||_*$

Значит это норма.

**Решение**  
Не является, т.к. не выплняется неравенство треугольника. Возьмем векторы $x, y = [1,2]$ тогда $||x + y||_? = 18 > ||x||_? + ||y||_? = 7+7=14$

**Задача 3**  
Выведите выражения для $||A||_1$ и $||A||_\infty$ через элементы матриц.  
*Подсказка: используя определение операторной нормы, получите оценку сверху для отношения, по которому берётся $\max$, а потом подберите элемент $x$ на котором оценка достигается.*

**Решение**  
По определению операторной нормы
$$
||A||_1 \ \sup_{x \ne 0} \frac{||A x||_1}{||x||_1}
$$
Оценим отношение под супремумом (кстати, вместо супремума можно писать максимум, т.к. можно брать sup по компакту - единичной сфере, а норма - непрерывная функция компонент вектора, следовательно достигает максимума)

$$
\frac{||A x||_1}{||x||_1} = \frac{\sum_i |\sum_j a_{ij} x_j |}{\sum_i |x_i|} \le \frac{\sum_i \sum_j |a_{ij}| | x_j |}{\sum_i |x_i|} = (\mbox{меняем суммы местами}) \\
\frac{\sum_j \left( | x_j | \sum_i |a_{ij}| \right)}{\sum_i |x_i|} \le (\mbox{все суммы по $i$ замеяем на максимальную}) \\
\frac{\max_{j}\sum_i |a_{ij}|  \sum_j | x_j |}{\sum_i |x_i|} = \max_{j}\sum_i |a_{ij}| 
$$
и можно подобрать вектор $x$, для которого в последнем переходе будет точное равенство. Это вектор вида
$$
x = \left[ \begin{array}{c} 
0 \\ \vdots \\ 1_{j^*} \\ \vdots \\ 0
\end{array}\right]
$$
где $j^*$ это индекса столбца, сумма модулей элементов в котором максимальна.

Для $||.||_\infty$ нормы ход решения такой же. Максимум будет достигаться на векторе вида:
$x = [sign(a_{i^* j}) ], \; j=1,n$
т.е. вектор, в котором стоят $\pm 1$ в зависимости от того, какой знак у соответствующего элемента строки

**Задача 4**  
Докажите, что 
$$
||A^T A||_1 \le ||A||_1 \; ||A||_\infty
$$

**Решение**
$||A^T A||_1 \le$ (свойство суб-ти матричной нормы) $ ||A^T||_1 ||A||_1$

Теперь остаётся заметить, что $||A^T||_1 = ||A||_\infty$, т.к. формулы для матричных норм $1$ и $\infty$  отличаются тем, что сумма берется по столбцу и по строке.

**Задача 6**  
Докажите, что для любой эрмитовой матрицы $H$ матрица $Q = (I - i H)^{-1} (I + i H)$ будет унитарной.

**Решение 1**  
Можно просто попытаться напрямую проверить, что $Q*Q = I$. Конкретные пути могут немного отличаться.

В выкладках используется, что операции $*,-1$ можно менять местами, и что когда берем  эти операции от произведения матриц, то нужно написать сомножители в обратном порядке, и к каждому применить операцию.
$$
Q* Q = (I - i H) (I + i H)^{-1} (I - iH)^{-1} (I + i H) = (I - i H) (I + H^2)^{-1}(I + i H) = 
$$
Чтобы доказить, что $A$ - единичная матрица, умножим выражение для неё слева на $I+iH$:
$$
(I+iH) A = (I + H^2)  (I + H^2)^{-1} (I + i H) = (I + i H)
$$
Следовательно $A = I$.

**Решение 2**  
Можно зайти издалека и вспомнить, что для нормальных матриц (частными случаями которых являются унитарные и эрмитовы матрицы) существует разложение (есть ортонормированный базис из собственных векторов):
$$
H = U \Lambda U^*
$$
$U$ - унитарная, $\Lambda$ -- диагональная. Кроме того, для эрмитовой матрицы все с.ч. на диагонали - действительные (это легко доказать из условия $H^* = H$).

Подставим это разложение в формулу для $Q$:
$$
Q = (I - iU \Lambda U^*)^{-1} (I + i U \Lambda U^*) = [U(I - i \Lambda)U^*]^{-1}  U (I + i \Lambda)U^* = U (I-i\Lambda)^{-1} U^* U (I + i \Lambda) U^* = U (I-i\Lambda)^{-1} (I + i \Lambda) U^*
$$

$$
(I-i\Lambda)^{-1} (I + i \Lambda) = D = diag\left(\frac{1 + i\lambda_k}{1 - i\lambda_k}\right)
$$
и видно, что $D^* = D^{-1}$

Тогда $Q* = U D^* U^* = Q^{-1}$ 



**Задача 7**  
Используя сингулярное разложение, докажите, что $\Vert A \Vert_2 = \sigma_{max}$.

**Решение:**  
$A = U \Sigma V^*$

$$
||A||_2 = \max_{||x||_2 = 1} ||A x||_2 = \max_{||x||_2 = 1} ||U \Sigma V^* x||_2 = (\mbox{Умножение на $U$ не меняет 2-норму}) \max_{||x||_2 = 1} ||\Sigma V^* x||_2 = (\mbox{переходим к $y = V^* x$}) = \max_{||y||_2 = 1} ||\Sigma y||_2 = \\
\max_{||y||_2 = 1} \left(\sum_i \sigma_i^2 y_i^2\right)^{1/2} \le \sigma_\max = \sigma_1 (\mbox{считаем, что упорядочили по убыванию})
$$
Максимум достигается на $y = [1, 0, \ldots, 0]^T$ и на $x = V y = v_1$

**Задача 9**  
Докажите, что 
$$
\Vert A \Vert_F = \sqrt{trace(A^* A)} = \sqrt{\sum_{k=1}^{\min\{m,n\}} \sigma_k^2(A)}
$$

**Решение**  
$A$ имеет размер $m \times n$
$$
trace(A^* A) = \sum_{i=1}^{n} [A^* A]_{ii} = \sum_{i=1}^{n} \left( \sum_{j=1}^{m} \overline{a}_{ji} a_{ji}\right) = \sum_{i=1}^{n}\sum_{j=1}^{m} |a_{ji}|^2
$$

Теперь докажем второе равенство.

$||A||_F^2 = \sum_{j=1}^n ||A(:,j)||_2^2 = \sum_{i=1}^{m} ||A(i,:)||_2^2$ (сумма квадратов 2-норм всех столбцов, или всех строк).

Отсюда следует, что умножение на унитарную матрицу слева или справа не меняет норму Фробениуса, т.к. каждый столбец или каждая строка в произведении имеют такую же 2-норму, как и в исходной матрице. 

Теперь запишем сингулярное разложение: 

$$
U^* A = \Sigma V^*
$$

$||\Sigma V^*||_F^2 = \sum_{j = 1}^{\min{m,n}} \sigma_j^2 ||V^*(j,:)||_2^2 =  \sum_{j = 1}^{\min{m,n}} \sigma_j^2$ 

Т.к. $V^*(j,:)$ - это столбцы унитарной матрицы и их 2-норма равна 1
